# Bivariate Data Analysis
## Bivariate data analysis is a statistical method used to examine the relationship between two variables. The goal of bivariate data analysis is to understand how changes in one variable are associated with changes in the other variable. In machine learning, bivariate data analysis is often used to evaluate the relationship between features (input variables) and target variables (outputs) in a dataset. In this we are doing analysis on two columns.

## Several techniques can be used for bivariate data analysis in machine learning, including:

*  ### Scatter plots: Scatter plots are used to visualize the relationship between two variables. In a scatter plot, each data point is represented as a dot, with the x-axis representing one variable and the y-axis representing the other. The pattern of the dots can reveal the strength and direction of the relationship between the two variables.
*  ### Correlation coefficients: Correlation coefficients are numerical measures of the strength of the relationship between two variables. The most commonly used correlation coefficient is Pearson's correlation coefficient, which ranges from -1 to 1 and indicates the strength and direction of the linear relationship between two variables.


*  ### Simple linear regression: Simple linear regression is a method used to model the relationship between two variables as a straight line. The goal of simple linear regression is to find the best-fitting line that summarizes the relationship between the two variables.

 ### These techniques can be used to gain insights into the relationships between variables in a dataset, and can help inform the development of machine learning models. By understanding the relationships between variables, machine learning algorithms can be designed to better capture the underlying patterns in the data and improve the accuracy of predictions.

## Heatmap

### A heatmap is a simple representation of data where values are represented as colors. Each value in the data is represented as a color in a grid, where each row and column represent variables. The color intensity represents the magnitude of the value in that cell.

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

### The color intensity in each cell of the heatmap represents the magnitude of the value in that cell. The darker the color, the higher the value, and vice versa. The rows and columns of the heatmap can represent different variables or dimensions, and the cells where they intersect represent the relationship between those variables.

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

## ClusterMap


### A clustermap, on the other hand, is a more advanced visualization that groups similar rows and columns together. It uses hierarchical clustering to rearrange the rows and columns of the data based on their similarity, so that similar values are placed together. The heatmap is then drawn for the reordered data, making it easier to identify patterns and relationships in the data.

### One key difference between a heatmap and a clustermap is that in a clustermap, the rows and columns of the data are reordered based on their similarity, making it easier to identify patterns and relationships in the data. For example, similar rows or columns will be grouped together, making it easier to see which variables tend to have similar values.



```python

sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

# Multivariate data analysis
## Multivariate data analysis is a statistical method used to examine the relationships among multiple variables. The goal of multivariate data analysis is to understand how changes in one variable are associated with changes in multiple other variables. In machine learning, multivariate data analysis is often used to evaluate the relationships among features (input variables) and target variables (outputs) in a dataset.

## There are several techniques that can be used for multivariate data analysis in machine learning, including:

*  ### Multiple regression: Multiple regression is a method used to model the relationship between multiple independent variables and a dependent variable. The goal of multiple regression is to find the best-fitting line that summarizes the relationships between the independent variables and the dependent variable.
*  ### Principal component analysis (PCA): PCA is a technique used to reduce the dimensionality of a dataset by transforming the variables into a set of uncorrelated variables called principal components. PCA can be used to identify the most important variables in a dataset and remove redundant or noisy variables.

*  ### Factor analysis: Factor analysis is a technique used to identify underlying factors that influence the relationships between variables in a dataset. Factor analysis is similar to PCA, but is used to identify the underlying factors that explain the relationships between variables rather than reducing the dimensionality of the data.
*  ### Cluster analysis: Cluster analysis is a technique used to group similar data points together into clusters. Cluster analysis can be used to identify patterns and relationships in a dataset and can be useful for discovering hidden structures in the data.

### These techniques can be used to gain insights into the relationships among variables in a dataset, and can help inform the development of machine learning models. By understanding the relationships among variables, machine learning algorithms can be designed to better capture the underlying patterns in the data and improve the accuracy of predictions

# Difference between Bivariate data analysis and Multivariate data analysis:
## Bivariate data analysis and multivariate data analysis are both statistical methods used to examine relationships between variables. The main difference between the two is the number of variables being analyzed.


*  ### Bivariate data analysis is focused on analyzing the relationship between two variables. It is used to understand how changes in one variable are associated with changes in another variable. Bivariate data analysis is typically used to evaluate the relationship between two features in a dataset or between a feature and the target variable.

*  ### Multivariate data analysis, on the other hand, is focused on analyzing the relationships among multiple variables. It is used to understand how changes in one variable are associated with changes in multiple other variables. Multivariate data analysis is typically used to evaluate the relationships among multiple features in a dataset or among multiple features and the target variable.

### In conclusion, bivariate data analysis is a subset of multivariate data analysis, and the latter is a more comprehensive method for analyzing relationships among variables. Both methods can be used in machine learning to gain insights into the relationships among variables in a dataset and to inform the development of machine learning models.
